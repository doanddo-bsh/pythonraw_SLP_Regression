```python
# import modul
import numpy as np
import csv
import time

np.random.seed(1234)
def randomize(): np.random.seed(time.time())
```

```python
# hyper parameter
# 가중치 파라미터 초기화 목적
RND_MEAN = 0 # 정규분포 난수값의 평균 
RND_STD = 0.0030 # 정규분포 난수값의 표준편차 

LEARNING_RATE = 0.001 # 학습률 
```

```python
input_cnt = 10
output_cnt = 1
```

실험용 메인 함수 정의

```python
# epoch_count 학습 횟수, mb_size 미니배치 크기, report 중간 보고 주기
def abalone_exec(epoch_count=10,mb_size=10, report=1):
    load_abalone_dataset() 
    init_model() # init model parameter
    train_and_test(epoch_count, mb_size, report) # train and test
```

```python
def load_abalone_dataset():
    with open(r'C:\Users\dcsun\Downloads\000_Study\014_python_raw_codding_dl_210103\Academy.ALZZA-master\Academy.ALZZA-master\data\chap01\abalone.csv') as csvfile:
        csvreader = csv.reader(csvfile)
        next(csvreader,None) # 첫 행을 읽지않고 건너뛰게함
        rows = []
        for row in csvreader:
            rows.append(row)
            
        global data, input_cnt, output_cnt

        # one-hot-encoding
        input_cnt, output_cnt = 10, 1
        data = np.zeros([len(rows),input_cnt+output_cnt])
        
        for n, row in enumerate(rows):
            if row[0] == 'I' : data[n,0] = 1
            if row[0] == 'M' : data[n,1] = 1
            if row[0] == 'F' : data[n,2] = 1
                
            # 이외 컬럼은 그대로 가지고 오기
            data[n, 3:] = row[1:]
    return data            
```

```python
def init_model():
    global weight, bias, input_cnt, output_cnt
    weight = np.random.normal(RND_MEAN, RND_STD, [input_cnt, output_cnt]) # row, col num가 input output 수
    bias = np.zeros([output_cnt]) # bias를 0으로
```

```python
# 74page
def train_and_test(epoch_count, mb_size, report):
    step_count = arrage_data(mb_size)
    test_x, test_y = get_test_data()
    
    for epoch in range(epoch_count):
        losses, accs = [], []
        
        for n in range(step_count):
            train_x, train_y = get_train_data(mb_size, n)
            loss, acc = run_train(train_x, train_y)
            print('loss : ', loss)
            losses.append(loss)
            accs.append(acc)
        print('mean loss : ', np.mean(losses))
        if report > 0 and (epoch+1) % report == 0 :
            acc = run_test(test_x, test_y) # 중간 평가 함수 
            # 손실도와 정확도의 평균을 출력
            print('Epoch {}: loss={:5.3f}, accuracy={:5.3f}/{:5.3f}'.format(epoch+1, np.mean(losses), np.mean(accs), acc))
            
    # 최종 평가 함수 
    final_acc = run_test(test_x, test_y)
    print('\nFinal Test: final accuray = {:5.3f}'.format(final_acc))
```

```python
# 학습 및 평가 데이터 획득 함수 정의
def arrage_data(mb_size):
    global data, shuffle_map, test_begin_idx
    shuffle_map = np.arange(data.shape[0])
    np.random.shuffle(shuffle_map)
    step_count = int(data.shape[0] * 0.8) # mb_size
    test_begin_idx = step_count * mb_size
    return step_count

def get_test_data():
    global data, shuffle_map, test_begin_idx, output_cnt
    test_data = data[shuffle_map[test_begin_idx:]]
    return test_data[:,:-output_cnt], test_data[:, -output_cnt:]

def get_train_data(mb_size, nth):
    global data, shuffle_map, test_begin_idx, output_cnt
    if nth == 0:
        np.random.shuffle(shuffle_map[:test_begin_idx])
    train_data = data[shuffle_map[mb_size*nth:mb_size*(nth+1)]]
    return train_data[:,:-output_cnt], train_data[:,-output_cnt:]
```

### run_train

```python
def run_train(x,y):
    output, aux_nn = forward_neuralnet(x) # 순전파 처리
    loss, aux_pp = forward_postproc(output, y) # 순전파 처리
    accuracy = eval_accuracy(output, y) # 보고용 정확도 계산
    
    G_loss = 1.0 
    G_output = backprop_postproc(G_loss, aux_pp) # 역전파 처리
    backprop_neuralnet(G_output, aux_nn) # 역전파 처리
    
    return loss,accuracy
```

## run_test

```python
def run_test(x,y):
    output, _ = forward_neuralnet(x)
    accuracy = eval_accuracy(output, y)
    return accuracy
```

### forward_neuralnet
#### 순전파 처리

```python
def forward_neuralnet(x):
    global weight, bias
    output = np.matmul(x,weight) + bias # x input에 weight을 곱하고 bias를 더해준다.
    return output, x
```

### backprop neuralnet - 역전파 처리

```python
def backprop_neuralnet(G_output, x):
    global weight, bias
    g_output_w = x.transpose()
    
    G_w = np.matmul(g_output_w, G_output)
    G_b = np.sum(G_output, axis = 0)
    
    weight -= LEARNING_RATE * G_w
    bias -= LEARNING_RATE * G_b
```

### 순전파 후처리

```python
def forward_postproc(output, y):
    diff = output - y
    square = np.square(diff)
    loss = np.mean(square)
    return loss, diff
```

### 역전파 후처리

```python
def backprop_postproc(G_loss, diff):
    shape = diff.shape
    
    g_loss_square = np.ones(shape) / np.prod(shape)
    g_square_diff = 2 * diff
    g_diff_output = 1
    
    G_square = g_loss_square * G_loss
    G_diff = g_square_diff * G_square
    G_output = g_diff_output * G_diff
    
    return G_output
```

### 정확도 계산 함수

```python
def eval_accuracy(output, y):
    mdiff = np.mean(np.abs((output - y)/y))
    return 1 - mdiff
```
